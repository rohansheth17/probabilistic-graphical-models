```python
import pandas as pd
import numpy as np
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, BatchNormalization, Activation, Reshape, UpSampling2D,Dense,Flatten, Dropout
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
import os
import cv2 as cv
import time
```

```python
img = os.listdir("TrainingSet")
image_values = []
path = "TrainingSet"
for file in os.listdir(path):
    ext = os.path.splitext(file)[1]
    temp = 255- cv.imread(os.path.join(path,file),0)
    image_values.append(np.array(temp).reshape(64,64,1))
    
data=pd.DataFrame(columns= ['image_id','values'])
data['image_id']= img
data['values']= image_values
```

```python
img1 = os.listdir("ValidationSet")
image_values1 = []
path = "ValidationSet"
for file in os.listdir(path):
    ext = os.path.splitext(file)[1]
    temp1 = 255- cv.imread(os.path.join(path,file),0)
    image_values1.append(np.array(temp1).reshape(64,64,1))

data1=pd.DataFrame(columns= ['image_id1','values1'])
data1['image_id1']= img1
data1['values1']= image_values1    
```

```python
training_data = pd.read_csv('dataset_seen_training_siamese.csv')
validation_data = pd.read_csv('dataset_seen_validation_siamese.csv')

samp1= pd.merge(training_data, data, left_on='left' , right_on='image_id')
samp2= pd.merge(samp1,data, left_on='right', right_on='image_id')
samp2=samp2.drop(['Unnamed: 0','image_id_x','image_id_y'],axis =1)
print(samp2)


```

```python
samp3= pd.merge(validation_data, data1, left_on='left', right_on='image_id1')
samp4= pd.merge(samp3,data1, left_on='right', right_on='image_id1')
samp4=samp4.drop(['Unnamed: 0','image_id1_x','image_id1_y'],axis =1)
print(samp4)
```

```python
imDim = 64
input_shape  = (imDim,imDim,1)
inp_img = Input(shape = input_shape, name = 'ImageInput')
model = inp_img


model = Conv2D(128, (5,5), activation='relu')(model)
model = MaxPooling2D()(model)
model = Dropout(0.2)(model)

model = Conv2D(128, (5,5), activation='relu')(model)
model = MaxPooling2D()(model)
model = Dropout(0.2)(model)

model = Flatten()(model)
model = Dropout(0.2)(model)
model = Dense(512, activation='relu')(model)


# #     model = Input(shape=(imDim,imDim,1))
# #     model.add(Input(shape = (imDim,imDim,1), name = 'FeatureNet_ImageInput'))
# model = Conv2D(32,kernel_size=(3, 3),activation='relu',input_shape=input_shape,padding='valid')(model)
# #     model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
# model = MaxPooling2D((2,2), padding='valid')(model)
# model = Conv2D(64, (3, 3), activation='relu',padding='valid')(model)
# #     model.add(Conv2D(32, (3, 3), activation='relu',padding='same'))
# model = MaxPooling2D((2,2),padding='valid')(model)
# #     model.add(Conv2D(16, (3, 3), activation='relu',padding='same'))
# model = Conv2D(128, (3, 3), activation='relu',padding='valid')(model)
# model = MaxPooling2D((2,2),padding='valid')(model)
# #     model.add(Conv2D(1, (3, 3), activation='relu',padding='same'))
# #     model.add(Conv2D(2, (3, 3), activation='relu',padding='same'))

# model = Conv2D(256, (1, 1), activation='relu',padding='valid')(model)
# model = MaxPooling2D((2,2),padding='valid')(model)
    
# model = Conv2D(64, (1, 1), activation='relu',padding='valid')(model)
# # model = MaxPooling2D((2,2),padding='valid')(model)
# model = Flatten()(model)

# # img_in = np.array((-1,imDim,imDim,1), dtype='float32')
# # img_in = tf.placeholder(shape=(imDim,imDim,1), dtype='float32')

feat = Model(inputs=[inp_img], outputs=[model],name = 'Feat_Model')
feat.summary()


# In[27]:

left_img = Input(shape = input_shape, name = 'left_img')
right_img = Input(shape = input_shape, name = 'right_img')


# In[28]:

left_feats = feat(left_img)
right_feats = feat(right_img)


# In[35]:

from keras.layers import concatenate
import random


# In[36]:

merged_feats = concatenate([left_feats, right_feats], name = 'concat_feats')
merged_feats = Dense(1024, activation = 'linear')(merged_feats)
merged_feats = BatchNormalization()(merged_feats)
merged_feats = Activation('relu')(merged_feats)
merged_feats = Dense(4, activation = 'linear')(merged_feats)
merged_feats = BatchNormalization()(merged_feats)
merged_feats = Activation('relu')(merged_feats)
merged_feats = Dense(1, activation = 'sigmoid')(merged_feats)
similarity_model = Model(inputs = [left_img, right_img], outputs = [merged_feats], name = 'Similarity_Model')
similarity_model.summary()
```

```python
samp5 = np.split(samp2,5)
val_labs = samp4['label']
left = list(samp4['values1_x'])
right = list(samp4['values1_y'])
test =[left,right]
```

```python
similarity_model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from keras.callbacks import TensorBoard
from keras.callbacks import EarlyStopping


num_epochs = 1
model_batch_size = 500
tb_batch_size = 500
early_patience = 500

tensorboard_cb   = TensorBoard(log_dir='logs', batch_size= tb_batch_size, write_graph= True)
earlystopping_cb = EarlyStopping(monitor='loss', verbose=1, patience=early_patience, mode='min')
for i in range(len(samp5)):
    lab = samp5[i]['label']
    x = list(samp5[i]['values_x'])
    y = list(samp5[i]['values_y'])  
    similarity_model.fit([x,y], lab , epochs=num_epochs , batch_size=model_batch_size 
                         , callbacks = [tensorboard_cb,earlystopping_cb])
```

```python
predicted_lab = similarity_model.predict(test)
```

```python
predicted_lab1 = []
for i in predicted_lab:
    predicted_lab1.extend(i)
predicted_final = []
for i in predicted_lab1:
    predicted_final.append(round(i))    

print(predicted_final)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(val_labs,predicted_final)
```

```python
def data_gen(data):
    mdata=np.zeros((len(data),2, 64, 64,1))
    y = np.zeros((len(data), 1))
    for i in range(len(data)):
        mdata[i,0,:,:,0] = cv.imread(path+data[i][0], 0)
        mdata[i,1,:,:,0] = cv.imread(path+data[i][1], 0)
        y[i] = data[i][2]     
    return mdata, y


def fill(df, data_type):
    pairRecords = []
    for i in df.index:
        pairRecords.append([df['left'][i], df['right'][i], df['label'][i]])
    return pairRecords

epochs = 5
data = pd.read_csv('dataset_seen_training_siamese.csv')
data = fill(data, data_type='training')

train_pairs, train_y = data_gen(data[:85793])
val_pairs, val_y = data_gen(data[85793:100092])
test_pairs, test_y = data_gen(data[100092:])


start = time.time()
similarity_model.fit([train_pairs[:, 0], train_pairs[:, 1]], train_y,
          batch_size=128,
          epochs=epochs,
          validation_data=([val_pairs[:, 0], val_pairs[:, 1]], val_y))
end=time.time()



y_pred_train = similarity_model.predict([train_pairs[:, 0], train_pairs[:, 1]])

y_pred_val = similarity_model.predict([val_pairs[:, 0], val_pairs[:, 1]])

y_pred_test = similarity_model.predict([test_pairs[:, 0], test_pairs[:, 1]])


accuracy_score(train_y,y_pred_train)
accuracy_score(val_y,y_pred_val)
accuracy_score(test_y,y_pred_test)
```
